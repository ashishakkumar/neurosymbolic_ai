```python
import numpy as np
from scipy.stats import ttest_rel
import pandas as pd

# F1-scores for each approach
talukder_2022 = np.array([0.62])
xia_2018 = np.array([0.60])
rotmensch_2017 = np.array([0.69])
in_context_llm = np.array([0.45])
pmkg = np.array([0.66])
pmmkg = np.array([0.67])
ragpmkg = np.array([0.68])

# Store the approaches and their corresponding F1-scores in a dictionary
approaches = {
    'Talukder 2022': talukder_2022,
    'Xia 2018': xia_2018,
    'Rotmensch 2017': rotmensch_2017,
    'In-Context LLM': in_context_llm,
    'PMKG': pmkg,
    'PMMKG': pmmkg,
    'RAGPMKG': ragpmkg
}

# Initialize lists to store the results
comparisons = []
t_statistics = []
p_values = []
significance = []

# Perform pairwise t-tests between each pair of approaches
approach_names = list(approaches.keys())
for i in range(len(approach_names)):
    for j in range(i + 1, len(approach_names)):
        approach_1 = approach_names[i]
        approach_2 = approach_names[j]
        f1_scores_1 = approaches[approach_1]
        f1_scores_2 = approaches[approach_2]
        
        # Paired t-test
        t_stat, p_val = ttest_rel(f1_scores_1, f1_scores_2)
        
        # Record the results
        comparisons.append(f"{approach_1} vs {approach_2}")
        t_statistics.append(t_stat)
        p_values.append(p_val)
        significance.append('Yes' if p_val < 0.05 else 'No')

# Create a DataFrame to store and display the results
results_df = pd.DataFrame({
    'Comparison': comparisons,
    't-statistic': t_statistics,
    'p-value': p_values,
    'Significant Difference (p < 0.05)': significance
})


results_df
```
